- 다익스트라 알고리즘을 정리해 보죠.
  1. `가장 가격이 싼 정점`, 즉 도달하는 데 시간이 가장 적게 걸리는 정점을 찾습니다. (`costs` 테이블을 통해서)
  2. `이 정점의 이웃 정점`에 대해 `현재 가격`보다 더 `싼 경로`가 존재하는지 확인 합니다. (`graph`해시 테이블과 `costs`값의 비교)
  3. 그래프 상의 모든 정점에 대해 `이러한 일을 반복합니다`.
  4. `최종 경로를 계산합니다`. (`parents` 테이블을 통해서)  

- 마지막으로 각 정점은 한 번씩만 처리해야 하므로 (`1단계를 무한으로 반복하지 않기 위해서`) 이미 처리한 정점을 추적하기 위한 배열(`processed`)도 필요 합니다.

```python
def find_lowest_cost_node(costs):
    lowest_cost = float("inf")
    lowest_cost_node = None
    for node in costs:        # 모든 정점을 확인합니다.
        cost = costs[node]    # 아직 처리하지 않은 정점 중 더 싼 것이 있으면, 
        if cost < lowest_cost and node not in processed:
            lowest_cost = cost  # 새로운 최저 정점으로 설정합니다.
            lowest_cost_node = node 
    return lowest_cost_node
```

```python
def dijkstra(graph, start, fin):
    node = find_lowest_cost_node(costs)   # 아직 처리하지 않은 가장 싼 정점을 찾습니다.
    while node is not None:         # 모든 정점을 처리하면 반복문이 종료합니다.
        cost = costs[node]
        neighbors = graph[node]
        for n in neighbors.keys():     # 모든 이웃에 대해 반복합니다.
            new_cost = cost + neighbors[n]
            if costs[n] > new_cost:    # 만약 이 정점을 지나는 것이 가격이 더 싸다면
                costs[n] = new_cost    # 정점의 가격을 갱신하고
                parents[n] = node       # 부모를 이 점점으로 새로 설정합니다.
        processed.append(node)         # 정점을 처리한 사실을 기록합니다.
        node = find_lowest_cost_node(costs)  # 다음으로 처리할 정점을 찾아 반복합니다.    
    
    # 4단계의 경로를 추측하기 위한 과정
    path = []
    current_node = fin
    while current_node != start:
        path.append(current_node)
        next_node = parents[current_node]
        current_node = next_node
    path.append(start)
    
    return costs[fin], path[::-1]
    
```
