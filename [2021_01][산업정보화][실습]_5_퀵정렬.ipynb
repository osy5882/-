## 1. 분할정복 농장 문제


- 우리는 분할정복으로 이 문제를 풀겠습니다.
  - 기본단계
  - 기본단계가 될때까지 나누는 단계

```
def maximum_square(a, b): # 가장 길이가 긴 정사각형, 모든 정사각형의 크기는 같아야하고.... 
  if a % b == 0: # 나누었을때 나머지를 계산하는 방법 ( // 연산 다르죠)
    return b
  elif b % a == 0:
    return a
  else:
    if a > b:
      a = a % b # 나누기 연산 후 몫이 아닌 나머지를 구함 (10 % 4 = 2)
    else:
      b = b % a
    return maximum_square(a, b)
```

```
maximum_square(1680, 640)
```

- 아래는 같은 문제를 푸는 더 효과적인 방법입니다. (상수 시간을 고려할때) 
- 반복문을 이용해서 분할정복 한 방법

```
def gcd(a, b): # 분할정복으로 문제해결
    mod = a % b
    while mod > 0:
        a = b
        b = mod
        mod = a%b
    return b
```

```
gcd(1680, 640)
```

```
gcd(640, 1680)
```


- 위의 코드를 재귀를 이용해서 푼 함수

```
def gcd_recursion(a, b):
    if a % b == 0: # 기본 단계
      return b
    a, b = b, a % b # 위의 반복문에서 사용한 변수의 값을 변경해주는 파이썬의 장점중 한개...
    return gcd_recursion(a,b) # 재귀 단계
```

```
gcd_recursion(640,1680)
```

## 2. 배열 더하기 예제

```
def sum(arr):
  total = 0
  for x in arr:
    total += x    # total = total + x 
  return total
```

```
sum([2,4,6])
```

```
def sum_recursion(arr):
  if len(arr) < 2: # 기본단계 (배열의 크기가 0 or 1)
    return arr[0]
  else:
    return arr[0] + sum_recursion(arr[1:]) # 재귀단계
```

## 3. 퀵정렬 

```
def quicksort(array):
  if len(array) < 2:  # 기본 단계: 원소의 개수가 0이거나 1이면 이미 정렬되어 있는 상태
    return array

  else:               # 재귀 단계
    pivot = array[0]  # 배열의 첫번째 값을 선택해서.. .

    less = []         # pivot보다 작은 배열
    greater = []      # pivot보다 큰 배열 
    for i in array[1:]: 
      if i <= pivot:  # 같은 값일때 왜 less에 넣어줘야 하는지. (질문: pivot과 중복된 값이 있다면?)
        less.append(i)
      else:
        greater.append(i)
    return quicksort(less) + [pivot] + quicksort(greater)
```

```
alist = [54,26,93,17,77,31,44,55,20] # 정렬되지 않은 배열
```

```
quicksort(alist) # 정렬된 배열
```

## 3.1 버블 정렬(The Bubble Sort), 참고



- 도서관에서 정렬되지 않은 책을 알파벳 순으로 정렬한다고 생각해 봅시다.
- 맨 왼쪽에 있는 책부터 시작하여 바로 오른쪽 책과 비교해보면서 만약 정렬이 안되있으면 위치를 바꾸면서 전체 서가를(오른쪽 부터 가장 큰 값 순으로) 정렬하는 방식을 생각해 봅시다.
- 이것이 버블 정렬 입니다. 

bubblepass.png

```
def bubbleSort(alist):
    for passnum in range(len(alist)-1,0,-1):   # range(9,0,-1) -> 9, 8, 7, 6, 5, 4, 3, 2, 1 
        for i in range(passnum):               # range(9) ... range(8) .... -> 
            if alist[i]>alist[i+1]:            # 왼쪽이 오른쪽보다 클때, 위치를 바꿔줘야함              
                # temp = alist[i]
                # alist[i] = alist[i+1]
                # alist[i+1] = temp
                alist[i], alist[i+1] = alist[i+1], alist[i]
```

```
alist = [54,26,93,17,77,31,44,55,20]
bubbleSort(alist)
print(alist)
```

## 3.3 병합정렬 (The Merge Sort), 참고


mergesortA.png

mergesortB.png

```
def merge_sort(list):
    if len(list) <= 1:
        return list

    mid = len(list) // 2
    leftList = list[:mid]
    rightList = list[mid:]
    
    leftList = merge_sort(leftList)
    rightList = merge_sort(rightList)
    return merge(leftList, rightList)

```

```
def merge(left, right):
    result = []
    while len(left) > 0 or len(right) > 0:
        if len(left) > 0 and len(right) > 0:
            if left[0] <= right[0]:
                result.append(left[0])
                left = left[1:]
            else:
                result.append(right[0])
                right = right[1:]
        elif len(left) > 0:
            result.append(left[0])
            left = left[1:]
        elif len(right) > 0:
            result.append(right[0])
            right = right[1:]
    return result
```

## 4. 선택정렬과 퀵정렬의 차이를 알아보자 (참고)

```
def find_smallest(arr):
    smallest = arr[0] # 가장 작은 정수를 저장
    smallest_index = 0 # 가장 작은 정수의 인덱스를 저장
    for i in range(1,len(arr)):
        if arr[i] < smallest:
            smallest = arr[i]
            smallest_index = i
    return smallest_index
```

```
def selection_sort(arr): # 입력값: 배열
    newArr = []
    for i in range(len(arr)):
        smallest = find_smallest(arr)  # 배열에서 가장 작은 정수를 찾아서 새로운 배열에 추가합니다.
        newArr.append(arr.pop(smallest))
    return newArr
```

```
def quicksort(array):
    if len(array) < 2:
        return array            # 기본 단계: 원소의 개수가 0이거나 1이면 이미 정렬되어 있는 상태
    
    else: 
        pivot = array[0]        # 재귀 단계
        
        less = []
        greater = [] 
        for i in array[1:]:    
            if i <= pivot:      # 기준 원소보다 작거나 같은 모든 원소로 이루어진 하위 배열
                less.append(i)
            else:               # 기준 원소보다 큰 모든 원소로 이루어진 하위 배열
                greater.append(i)
                
        return quicksort(less) + [pivot] + quicksort(greater)
```
