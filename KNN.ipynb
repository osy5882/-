```python
#거리 제공 함수 정의
import numpy as np
import random
def distance(p1,p2):
    return np.sqrt(np.sum(np.power(p2-p1,2)))
```

```python
#과반수 투표 함수 정의
def majority_vote(votes): 
    vote_counts = {}
    for vote in votes:
        if vote in vote_counts:
            vote_counts[vote] += 1
        else:
            vote_counts[vote] = 1
    most = []
    max_count = max(vote_counts.values())
    for vote,count in vote_counts.items():
        if count == max_count:
            most.append(vote)
    return random.choice(most)
```

```python
votes = [1,2,3,1,2,3,1,2,3,3,3,3]
print(majority_vote(votes))
```

```python
#KNN의 처리 과정
#1. 전체 데이터 집단 확인
#2. 그중 한점과 다른 점간의 거리들을 구한다
#3. 그 거리를 나열해 기준에 적합한 점들을 반환해준다.

#이웃 찾기 함수 정의
def find_nearest_neighbors(p, points, k = 5):
    distances = np.zeros(points.shape[0])
    for i in range(len(distances)):
        distances[i] = distance(p, point[i])
    ind = np.argsort(distances)
    return ind[:k]
```

```python
def knn_predict(p,points,outcomes, k= 5):
    ind = find_nearest_neighbors(p,point,k)
    return majority_vote(outcomes[ind])

points = np.array([[1,1],[2,1],[2,2],[1,2],[3,2]])
```
